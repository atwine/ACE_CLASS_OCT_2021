Identifying and removing outliers displayed using the IQR (Interquartile range).
This will be the difference between the third quartile 75th(upper)or Q3 and 25th (lower) or Q1 as displayed in the box plot.
IQR = Q3 - Q1.
It is a measure of the dispersion similar to standard deviation or variance, but is much more robust against outliers.

```python
Q1 = heart_df.quantile(0.25)
Q3 = heart_df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

Comment:

```python
print(heart_df < (Q1 - 1.500 * IQR)) |(heart_df > (Q3 + 1.500 * IQR))

import warnings
warnings.filterwarnings('ignore')
```

False means presence of the outliers
True indicates absence of outliers
